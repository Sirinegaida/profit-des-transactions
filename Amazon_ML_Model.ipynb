```python
import pandas as pd
data = pd.read_csv("sales_returns.csv")
data.head()
```

```python
data.info()
```

```python
X = data.loc[:,["cust_age","cust_gender","city","sub_category"]]
y = data.returned
X.head()
```

```python
from sklearn.preprocessing import LabelEncoder
enc = LabelEncoder()
X.cust_gender = enc.fit_transform(X.cust_gender)
X.sub_category = enc.fit_transform(X.sub_category)
X.city = enc.fit_transform(X.city)
X.head()
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier()
model.fit(X,y) # training, modeling, learning, fitting
```

```python
model.predict([[27,0,156,2]]) # ship this order
```

```python
model.predict([[20,1,100,13]]) # Check before shipping
```

```python
model.score(X,y)
```
